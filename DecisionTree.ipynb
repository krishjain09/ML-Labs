```python
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
#Laod datasets
iris = load_iris()
```

```python
X = iris.data
Y = iris.target
```

```python
#X_train => Features(input data) used to train the model
#Y_train => labels(Correct Answer) for training data
#X_test ==> Features(unseen data) to test the model
#Y_test => Label (Correct Answer) of training_data

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
#Create Decision Tree model
clf = DecisionTreeClassifier(criterion='entropy',random_state=0)
```

```python
#Train the model
clf.fit(X_train,Y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("Accuracy Matrix", accuracy_score(Y_test,y_pred))
```

```python
print("Confusion Matrix", confusion_matrix(Y_test,y_pred))
```

```python
print("Classification Matrix", classification_report(Y_test,y_pred))
```

```python
plt.figure(figsize=(12,8))
plot_tree(clf,filled=True,class_names=iris.target_names,feature_names=iris.feature_names)
plt.show()
```
